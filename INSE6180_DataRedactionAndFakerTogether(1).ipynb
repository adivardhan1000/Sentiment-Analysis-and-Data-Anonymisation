```python
import spacy
import pandas as pd
import time
from faker import Faker

check = spacy.prefer_gpu()
print(check)

import torch
if torch.cuda.is_available():
    print("GPU is available")
else:
    print("GPU is not available")

# Load the spaCy model
nlp = spacy.load("en_core_web_sm")

# Create a Faker instance
faker = Faker()

past_time = time.time()

def process_text(text):
    doc = nlp(text)
    entities = []
    for ent in doc.ents:
        entities.append((ent.start_char, ent.end_char, ent.label_, ent.text))
    return entities

def redact_entities(text, entities, redaction_token="[REDACTED]"):
    """Replaces entities in the text with a redaction token."""
    start_indices = [ent[0] for ent in entities]
    end_indices = [ent[1] for ent in entities]
    result = ""
    last_end = 0
    for start, end in zip(start_indices, end_indices):
        result += text[last_end:start] + redaction_token
        last_end = end
    result += text[last_end:]  # Add remaining text
    return result

def update_entities(text, entities, replacement_token="[FAKE_VALUE]"):
    """Replaces entities in the text with a replacement token."""
    start_indices = [ent[0] for ent in entities]
    end_indices = [ent[1] for ent in entities]
    result = ""
    last_end = 0
    for start, end, entity_label in zip(start_indices, end_indices, [ent[2] for ent in entities]):
        if entity_label == "PERSON":
            fake_value = faker.name()
        elif entity_label == "GPE":
            fake_value = faker.city()
        elif entity_label == "LOC":
            fake_value = faker.country()
        elif entity_label == "DATE":
            fake_value = faker.date()
        elif entity_label == "ORG":
            fake_value = faker.company()
        elif entity_label == "EMAIL":
            fake_value = faker.email()
        elif entity_label == "PHONE":
            fake_value = faker.phone_number()
        else:
            fake_value = replacement_token

        result += text[last_end:start] + fake_value
        last_end = end
    result += text[last_end:]  # Add remaining text
    return result


# Specify the path to your CSV file (adjust accordingly)
data_path = "data.csv"
results_file = "entity_analysis_results2.txt"
redacted_data_file = "redacted_data.csv"
updated_data_file = "updated_data.csv"

def update_results_file(stats):
    with open(results_file, "a") as f:  # Append to the results file
        f.write(f"Checkpoint: {stats['checkpoint']}\n")
        f.write(f"Total Entities: {stats['total_entities']}\n")
        f.write(f"Rows with Entities: {stats['unique_rows']}\n")
        f.write("-------------------------\n")

# Read the CSV file
df = pd.read_csv(data_path, encoding="latin_1", names=['target', 'ids', 'date', 'flag', 'user', 'text'])

# Entities to search for
target_entities = ["PERSON", "GPE", "LOC", "DATE", "ORG", "EMAIL", "PHONE"]
stats = {
    'unique_rows': 0,
    'total_entities': 0,
    'checkpoint': 0
}

# Read the last checkpoint value
try:
    with open(results_file, "r") as f:
        lines = f.readlines()
        if lines:
            stats['checkpoint'] = int(lines[-4].split(":")[-1].strip())
            stats['total_entities'] = int(lines[-3].split(":")[-1].strip())
            stats['unique_rows'] = int(lines[-2].split(":")[-1].strip())
except FileNotFoundError:
    pass

df_redacted = df.copy()
df_updated = df.copy()

batch_size = 10000
print(f"Total Processing Time: {time.time() - past_time:.3f} seconds")
past_time = time.time()
# Process the rows starting from the last checkpoint
for i in range(stats['checkpoint'] * batch_size + 1, len(df), batch_size):
    batch_df = df.iloc[i:i+batch_size]
    batch_redacted = batch_df.copy()
    batch_updated = batch_df.copy()

    for idx, row in batch_df.iterrows():
        text = row['text']
        entities = process_text(text)

        if entities:
            stats['unique_rows'] += 1
            stats['total_entities'] += len(entities)

            # Redact entities
            batch_redacted.at[idx, 'text'] = redact_entities(text, entities)

            # Update entities with fake values
            batch_updated.at[idx, 'text'] = update_entities(text, entities)

    df_redacted = pd.concat([df_redacted, batch_redacted])
    df_updated = pd.concat([df_updated, batch_updated])

    stats['checkpoint'] = i // batch_size
    update_results_file(stats)
    print(f"Processed up to index {i}")

    # Save the redacted batch to the redacted data file
    batch_redacted.to_csv(redacted_data_file, mode='a', header=False, index=False, encoding='latin_1')

    # Save the updated batch to the updated data file
    batch_updated.to_csv(updated_data_file, mode='a', header=False, index=False, encoding='latin_1')

    print(f"Total Processing Time: {time.time() - past_time:.3f} seconds")
    past_time = time.time()

# Save the final redacted and updated DataFrames
df_redacted.to_csv(redacted_data_file, mode='w', header=True, index=False, encoding='latin_1')
df_updated.to_csv(updated_data_file, mode='w', header=True, index=False, encoding='latin_1')

print(f"Total Processing Time: {time.time() - past_time:.3f} seconds")
print("Redacted and updated data saved successfully.")
```
